```python
# importing libraries

import cv2
import matplotlib.pyplot as plt
```

```python
# loading and plotting the original image

img = cv2.imread("sunflower.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)    #converting image from BGR to RGB
plt.figure(figsize = (7,7))
plt.imshow(img)
plt.title("Original Image")
plt.show()
```

```python
# converting original image to grayscale

img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize = (7,7))
plt.imshow(img_gray, cmap ="gray")
plt.title("GrayScale Image")
plt.show()
```

```python
# Inverting the image

img_invert = cv2.bitwise_not(img_gray)
plt.figure(figsize = (7,7))
plt.imshow(img_invert, cmap = "gray")
plt.title("Inverted Image")
plt.axis("off")
plt.show()
```

```python
# smoothing the image

img_smoothing = cv2.GaussianBlur(img_invert, (23,23), sigmaX =0, sigmaY =0)   
plt.figure(figsize = (7,7))
plt.imshow(img_smoothing, cmap="gray")
plt.title("Blur Image")
plt.axis("off")
plt.show()
```

```python
# conveting image to pencil sketch

final_img = cv2.divide(img_gray, 255-img_smoothing, scale = 255)
plt.figure(figsize =(7,7))
plt.imshow( final_img , cmap ="gray")
plt.title("Final Sketch Image")
plt.axis("off")
plt.show()
```

```python
# The Final Output
# using subplot to display all the images in one frame

plt.figure(figsize =(18,18))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")

plt.subplot(1,5,2)
plt.imshow(img_gray, cmap ="gray")
plt.axis("off")
plt.title("GrayScale Image")

plt.subplot(1,5,3)
plt.imshow(img_invert, cmap ="gray")
plt.axis("off")
plt.title("Inverted Image")

plt.subplot(1,5,4)
plt.imshow(img_smoothing, cmap ="gray")
plt.axis("off")
plt.title("Blur Image")

plt.subplot(1,5,5)
plt.imshow(final_img, cmap ="gray")
plt.axis("off")
plt.title("Final Sketch Image")
plt.show()
```
